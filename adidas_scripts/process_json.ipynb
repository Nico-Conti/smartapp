### Process Title and Category

```python
import sys
import os

import regex as re

curr_dir = os.getcwd()

print("Current Directory:", curr_dir)

parent_dir = os.path.abspath(os.path.join(curr_dir, '..'))

print("Parent Directory:", parent_dir)

sys.path.append(parent_dir)

import json

group = 'men-sneakers'

with open(f'{parent_dir}/adidas_catalog/uomo/{group}.json', 'r', encoding='utf-8') as f:
    data = json.load(f)
```

### Process Image for Emedding

```python

count = 0
for item in data:
    count += 1
    # print(count)

    if item['images']:
        rescale_img = re.sub(r'w_383,h_383', 'w_766,h_766', item['images'])
        item['image_link'] = rescale_img

    else:
        print("No images found for item ID:", item.get('id', 'N/A'))
        break

```

```python
with open(f'{parent_dir}/adidas_scripts/{group}.json', 'w', encoding='utf-8') as f:
    json.dump(data, f, indent=4, ensure_ascii=False)
```
