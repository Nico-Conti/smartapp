### Process Title and Category

```python
import sys
import os

import regex as re

curr_dir = os.getcwd()

print("Current Directory:", curr_dir)

parent_dir = os.path.abspath(os.path.join(curr_dir, '..'))

print("Parent Directory:", parent_dir)

sys.path.append(parent_dir)

import json

group = 'jackets-coats'

with open(f'{parent_dir}/h&m_scripts/processed_json/{group}.json', 'r', encoding='utf-8') as f:
    data = json.load(f)
```

### Process ID extraction SKU

```python
for item in data:
    url = item.get('url', '')
    match_id = re.search(r'(\d+)\.html$', url)

    if match_id:
        item['id'] = match_id.group().replace('.html', '').strip()
```

## Add unisex Audience

```python
for item in data:
    # setdefault() checks for the key 'audience'.
    # If 'audience' is missing, it creates the key and sets its value to 'unisex'.
    # If 'audience' is present, it does nothing to the dictionary.
    item.setdefault('audience', 'unisex')
```

### Process Image for Emedding

```python

count = 0
for item in data:
    count += 1
    # print(count)

    if item['images']:

        item['image_link'] = item['images'][0]

        item['image_link'] = item['image_link'].split('?')[0]
        
    else:
        print("No images found for item ID:", item.get('id', 'N/A'))
        break
```
